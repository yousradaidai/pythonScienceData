```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_predictions = pd.read_csv('../DATA/sales_predictions.csv')
sales_predictions.info()
```

```python
sns.scatterplot(data=sales_predictions, y="item_price", x="date")
```

```python
new_sales_predictions = sales_predictions.sample(n=50)
new_sales_predictions.info()
```

```python
sns.scatterplot(data=new_sales_predictions, y="item_price", x="date")
```

```python
sns.lineplot(data=new_sales_predictions, y="item_price", x="date")
```

```python
sns.lineplot(data=new_sales_predictions.query('item_price<2000'), y="item_price", x="date")
```

```python
new_sales_predictions['date'] = pd.to_datetime(new_sales_predictions['date'])
new_sales_predictions.info()
```

```python
sns.lineplot(data=new_sales_predictions, y="item_price", x="date")
```

```python
sns.lineplot(data=new_sales_predictions.query('item_price<2000 and date>20130126'), y="item_price", x="date")
```
